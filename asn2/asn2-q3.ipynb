```python
# boilerplate imports

import matplotlib.pyplot as plt
import skimage.data as skdata
import skimage.feature as skfeat
import skimage.color as skcolor
import skimage.io as skio
```

# Step 1: Perform Canny edge detection on the `skdata.astronaut()` image.

- call skimage.astronaut() to obtain the image
- convert it to grayscale
- perform Canny edge detection (`skfeat.canny()` function) on the grayscale image -- select the parameters to optimize the edge detection to find only the edges of major salient objects in the scene such as the person, helmet, flag, and space shuttle model (you won't be able to do this perfectly, but do the best you can).  
- display the result of the Canny edge detection as a binary image.

```python
# Write your code here.

img = skdata.astronaut()
gray = skcolor.rgb2gray(img)

edges = skfeat.canny(gray, sigma=2.0, low_threshold=0.1, high_threshold=0.25)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title('Original: astronaut')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(edges, cmap='gray')
plt.title('Canny edges')
plt.axis('off')
plt.show()
```

# Step 2: Justify your parameter selection

Explain why you selected the parameters that you did and justify why they are appropriate for the given edge detection task and context.

*Write your answer here:* Since we only want to detect the major edges in this image, we therefore want higher smoothing done on the image along
with a larger high threshold, which will prioritize major edge detection (large intensity changes) through more smoothing (2.0), along with a higher threshold, so only edges that stronger than 25% of the maximum gradient (0.25). Having a higher value for the lower threshold also helps with this, as it makes the edge detection less sensitive, which is what we want (0.1).


# Step 3: Repeat for the `mushroom.jpg` image.

- load the image file `mushroom.jpg` provided with this assignment and convert it to grayscale
- perform Canny edge detection on the grayscale image -- select the parameters to find only the edges of the five mushrooms in the foreground that are in focus (you won't be able to do this perfectly, but do the best you can).
- display the result of the Canny edge detection as a binary image.

```python
# Since this image is a little larger, it may display with a fuzzy appearance if you don't increase the 
# resolution of figure rendering a bit.  Before you call plt.imshow(), call plt.figure(dpi=300).  
# This will result in a clearer display of the binary edge image.

img = skio.imread('mushroom.jpg')
gray = skcolor.rgb2gray(img)

edges = skfeat.canny(gray, sigma=1.0, low_threshold=0.2, high_threshold=0.45)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title('Original: mushroom')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(edges, cmap='gray')
plt.title('Canny edges')
plt.axis('off')
plt.show()
```

# Step 4: Justify your parameter selection

Explain why you selected the parameters that you did and justify why they are appropriate for the given edge detection task and context.

*Write your answer here:* For this image, I chose a lower value for sigma, since the image foreground is fairly similar to the background, so too much smoothing would blur the front 5 mushrooms too much. However, I still had to keep the upper threshold high and the lower threshold high, since the 5 front mushrooms and neighbouring ones would have similar edge intensity, therefore the lower threshold had to be raised to reduce the noise from neighboring mushrooms, and the upper threshold had to be high to maximize the intensity different between the 5 mushrooms and neighboring mushrooms.
